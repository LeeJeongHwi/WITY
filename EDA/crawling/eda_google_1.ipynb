```python
# 컬럼명 변경
df.rename(
    columns={
        "Number": "id",
        "Name": "name",
        "Address": "address",
        "Keyword": "category",
        "Rating": "rating",
        # "영업시간": "business_hours"
        # "Review": "reviews"
        "Price_per_one": "price_per_one"
    },
    inplace=True
)
print(df)
```

```python
# Reviewer 컬럼 삭제
df = df.drop(columns=["Domain"])

# domain 컬럼 추가
df["domain"] = "google"
df["business_hours"] = ""

# 컬럼 순서 재배열
# columns_order = ["id", "domain", "name", "reviews"]
columns_order = ["id", "domain", "name", "category", "rating", "address", "business_hours", "price_per_one"]
df = df[columns_order]

# 결과 확인
print(df)
```

```python
# 'rating' 열에서 숫자만 남기기
df['rating'] = df['rating'].str.extract('(\d+\.?\d*)').astype(float)
df
```

```python
# 행이 동일한 중복된 행 찾기(크롤링 시간에 따라 영업시간 다름 이슈로 id 기준 중복 제거하기로 함)
# print(len(df[df.duplicated()]))
# print(df[df.duplicated()])

# 눈으로 확인하고 싶어서~
# duplicates = df[df.duplicated()]
# duplicates.to_csv("duplicates.csv", index=False)

# 중복 데이터 제거
# df = df.drop_duplicates()
# print(df.shape)
```

```python
# 중복된 id를 가진 행들 찾기
duplicates = df[df.duplicated(subset=["id"], keep="first")]
# 개수 출력
print(len(duplicates))
# id 중복 행 제거
# df = df.drop_duplicates(subset=['id'], keep='first')
# print(df.shape)
```

```python
max_id = 6605 # 6605, 377
existing_ids = df['id'].dropna().astype(int).tolist()
all_ids = list(range(1, max_id + 1))

# 2. 기존 ID를 제외한 사용 가능한 ID 생성
available_ids = [i for i in all_ids if i not in existing_ids]

# 3. 새로운 데이터프레임 만들기 (1부터 max_id까지 모든 ID 포함)
full_df = pd.DataFrame({'id': all_ids})

# 4. 기존 데이터와 병합하여 기존 ID 데이터 유지
full_df = full_df.merge(df, on='id', how='left')

# 5. 도메인 값이 비어 있는 경우 'google'로 채우기
full_df['domain'] = full_df['domain'].fillna('google')

# 6. 결과 확인
print(full_df)
```

```python
for col in ['domain', 'name', 'category', "address", "business_hours", "price_per_one"]:
    full_df[col] = full_df[col].fillna('')
print(full_df)
```

```python
# 빈 문자열이나 None을 NaN으로 변경
full_df.replace({"": np.nan}, inplace=True)
# 결과 확인
print(full_df)
```

```python
# csv 저장
full_df.to_csv("google_store_info.csv", index=False)
```
