```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# store reviews 정제 부분
reviews_pre = pd.read_csv("all_store_reviews_3.csv")
print(reviews_pre.shape)
reviews_post = pd.read_csv("all_store_reviews_4.csv")
print(reviews_post.shape)
```

```python
# 글자 수 계산
reviews_pre["reviews_pre_length"] = reviews_pre["reviews"].str.len()
reviews_post["reviews_post_length"] = reviews_post["reviews"].str.len()
print(reviews_pre["reviews_pre_length"].max())
print(reviews_post["reviews_post_length"].max())

# 히스토그램 비교
max_length = 1940  # 최댓값(1934)
bin_size = 10  # 10 단위로 나눔
plt.figure(figsize=(12, 10))
sns.histplot(reviews_pre["reviews_pre_length"], bins=range(0, 600, bin_size), 
             color="blue", label="Before", kde=False, alpha=0.5)
sns.histplot(reviews_post["reviews_post_length"], bins=range(0, 600, bin_size), 
             color="red", label="After", kde=False, alpha=0.5)

# x축, y축 라벨 및 제목 설정
plt.xlabel("Review Length")
plt.ylabel("Frequency")
plt.title("Comparison of Review Length Before and After Cleaning")
plt.xticks(range(0, 600 + 1, 100))  # x축 눈금 100 단위로 설정
plt.legend()
plt.show()
```

```python
# spot category 분류 부분
spot_pre = pd.read_csv("naver_spot_info.csv")
print(spot_pre.shape)
spot_post = pd.read_csv("all_spot_info_4.csv")
print(spot_post.shape)
```

```python
# 고유한 카테고리 값 찾기
categories_pre = spot_pre['category'].unique()
categories_post = spot_post['category'].unique()

# 고유 카테고리 값 출력
print("정제 전 카테고리 종류:")
print(len(categories_pre))
print(categories_pre)

print("정제 후 카테고리 종류:")
print(len(categories_post))
print(categories_post)

# 카테고리 차이 비교
categories_diff_pre = set(categories_pre) - set(categories_post)  # 정제 전 카테고리 중 후에 없는 것
categories_diff_post = set(categories_post) - set(categories_pre)  # 정제 후 카테고리 중 전에는 없는 것

print("정제 전에는 있었지만 정제 후에는 없는 카테고리:")
print(categories_diff_pre)

print("정제 후에 새로 추가된 카테고리:")
print(categories_diff_post)
```

```python
# 'main_category'에 따른 고유한 'category' 값 확인
category_by_main_category = spot_post.groupby('main_category')['category'].unique()

# 결과 출력
print("main_category별 고유한 category 종류:")
print(category_by_main_category.to_string())
```

```python
# store info data 분포
df = pd.read_csv("all_store_info_5.csv")
print(df.shape)
print(df)

# 'main_category'별로 'category'의 개수 세기
category_count_by_main_category = df.groupby('main_category')['category'].nunique()
print(category_count_by_main_category)

# 그래프 그리기
plt.figure(figsize=(5, 4))  # 그래프 크기 설정
category_count_by_main_category.plot(kind='bar', color='orange')  # 막대 그래프
plt.title('Number of Categories')  # 그래프 제목
plt.xlabel('Main Category')  # x축 레이블
plt.ylabel('Number of Categories')  # y축 레이블
# plt.xticks(rotation=45, ha='right')  # x축 레이블 회전
plt.xticks(rotation=0)  # 글씨 기울기를 0도로 설정

# x축 이름 변경
new_labels = ['Restaurant', 'Cafe', 'Pub']  # 기존 이름에 맞게 새 이름 리스트 작성
plt.xticks(ticks=range(len(new_labels)), labels=new_labels)  # 새 레이블 적용

plt.tight_layout()  # 레이아웃 자동 조정
plt.show()
```
